```python
# слишком много памяти
import sys
sys.setrecursionlimit(20000)

def profit(prices, N, K):
    prof = 0
    
    if N == 2:
        return prices[1] - prices[0]
    else:
        prof = profit(prices[1:], N - 1, K)
        
    for i in range(1, N):
        if i <= K and prices[i] - prices[0] > prof:
            prof = prices[i] - prices[0]
    
    return prof 

N, K = map(int, input().split())
prices = list(map(int, input().split()))

profit_new = profit(prices, N, K)
if profit_new < 0:
    profit_new = 0

print(profit_new)
```
