```python
# Проверка на квадрат

class Point:
	def __init__(self, x, y):
		self.x = x
		self.y = y

def distSq(p, q):
	return (p.x - q.x) * (p.x - q.x) +\
		(p.y - q.y) * (p.y - q.y)

def isSquare(p1, p2, p3, p4):

	d2 = distSq(p1, p2) # from p1 to p2
	d3 = distSq(p1, p3) # from p1 to p3
	d4 = distSq(p1, p4) # from p1 to p4

	if d2 == 0 or d3 == 0 or d4 == 0: 
		return False

	if d2 == d3 and 2 * d2 == d4 and \
					2 * distSq(p2, p4) == distSq(p2, p3):
		return True

	if d3 == d4 and 2 * d3 == d2 and \
					2 * distSq(p3, p2) == distSq(p3, p4):
		return True

	if d2 == d4 and 2 * d2 == d3 and \
					2 * distSq(p2, p3) == distSq(p2, p4):
		return True

	return False

if __name__=="__main__":
	p1 = Point(-1, 1)
	p2 = Point(1, 1)
	p3 = Point(-1, -1)
	p4 = Point(1, -1)
	
	if isSquare(p1, p2, p3, p4):
		print('Yes') 
	else:
		print('No')
```

```python
def res_ij(all_points, c1, c2):
	res = 0
	result = []
	if c1 not in all_points:
		res += 1
		result.append(c1)

	if c2 not in all_points:
		res += 1
		result.append(c2)
	return res, result

def var1(all_points, x1, y1, x2, y2):
	x11 = x1 + (y2 - y1)
	y11 = y1 - (x2 - x1)

	x22 = x2 + (y2 - y1)
	y22 = y2 - (x2 - x1)

	c1=(x11, y11)
	c2=(x22, y22)

	return  res_ij(all_points, c1, c2)

def var2(all_points, x1, y1, x2, y2):
	x11 = x1 - (y2 - y1)
	y11 = y1 + (x2 - x1)

	x22 = x2 - (y2 - y1)
	y22 = y2 + (x2 - x1)
 
	c1=(x11, y11)
	c2=(x22, y22)

	return  res_ij(all_points, c1, c2)


def var3(all_points, x1, y1, x2, y2):
	is_correct_int = lambda x: (x // 2) * 2 == x
	if x1 < x2:
		yr = y2
		yl = y1
	else:
		yl = y2
		yr = y1
	if yr > yl:
		x11_2 = (x1 + x2) - abs(y2 - y1)
		y11_2 = (y1 + y2) + abs(x2 - x1)

		x22_2 = ((x1 + x2) + abs(y2 - y1))
		y22_2 = (y1 + y2) - abs(x2 - x1)
	else:
		x11_2 = (x1 + x2) - abs(y2 - y1)
		y11_2 = (y1 + y2) - abs(x2 - x1)

		x22_2 = ((x1 + x2) + abs(y2 - y1))
		y22_2 = (y1 + y2) + abs(x2 - x1)
	if all(is_correct_int(x) for x in [x11_2, y11_2, x22_2, y22_2]):
		c1=(x11_2 // 2, y11_2 // 2)
		c2=(x22_2 // 2, y22_2 // 2)
		return res_ij(all_points, c1, c2)
	return  None, None

def main(n, table) -> list:
	if n == 1:
		(x,y) = table[0]
		return [3, {
			(x+1, y),
			(x, y-1),
			(x+1, y+1),
		}]
	all_points = set(tuple(x) for x in table)
	res = 4
	result = []
	for i, (x1, y1) in enumerate(table[:-1]):
		for j, (x2, y2) in enumerate(table[i+1:]):

			res_ij, result_ij = var1(all_points, x1, y1, x2, y2)
			if res_ij < res:
				res = res_ij
				result = result_ij

			res_ij, result_ij = var2(all_points, x1, y1, x2, y2)
			if res_ij < res:
				res = res_ij
				result = result_ij

			res_ij, result_ij = var3(all_points, x1, y1, x2, y2)
			if res_ij is not None and res_ij < res:
				res = res_ij
				result = result_ij

			if res == 0:
				return  [res, set(result)]
	return [res, set(result)]

n = int(input())
list_n = []
for i in range(n):
    list_n .append(tuple(map(int, input().split())))

n, res = main(n, list_n )
print(n)
for line in res:
    print(*line)
```
