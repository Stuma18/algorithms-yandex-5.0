```python
import json 

def apply(phrases, transformation_rule): 
    if transformation_rule == "10": 
        return sorted([phrase for phrase in phrases if vowels(phrase)], reverse=True) 
    elif transformation_rule == "20": 
        return sorted([phrase for phrase in phrases if length(phrase)], reverse=True) 
    elif transformation_rule == "30": 
        return sorted(phrases, reverse=True) 
    else: 
        return [] 

def read(name): 
    with open(name, 'r') as f: 
        return json.load(f) 
 
def vowels(phrase): 
    vowels = "aeiouy" 
    distinct_vowels = set([char for char in phrase if char in vowels]) 
    return len(distinct_vowels) >= 2 
 
def length(phrase): 
    return len(phrase) % 2 == 0 


name = input().strip() 
rules = read(name) 

output_data = {} 
line_number = 1 
try: 
    while True: 
        line = input().strip() 
        if line == "": 
            break 
        phrases = line.split("_") 
        rule = rules.get(str(line_number), "30")
        output_data[str(line_number)] = apply(phrases, rule) 
        line_number += 1 
except EOFError: 
    pass 

with open('output.json', 'w') as outfile: 
    json.dump(output_data, outfile, ensure_ascii=False, indent=2) 
```

```python
from enum import Enum
import json


class ParseType(Enum):
    """Enumeration of parsing types."""
    MIN_TWO_GLAS = "10"
    CHET_LEN = "20"
    USE_ALL = "30"


GLAS = {"a", "e", "u", "i", "o", "y"}

file_name = input()
FILE_STRUCTURE = None

with open(file_name, "r", encoding='utf-8') as f:
    FILE_STRUCTURE = json.loads(f.read())


def filter_phrases_with_min_two_different_glas(phrases):
    """Filter phrases with at least two different vowels."""
    for phrase in phrases:
        glas_count = {}
        for letter in phrase:
            if letter in GLAS:
                if letter in glas_count:
                    glas_count[letter] += 1
                else:
                    glas_count[letter] = 1
                if len(glas_count) >= 2:
                    yield phrase
                    break


output_json = {}
i = 1

while data := input():
    data_phrases = data.split("_")
    parse_type = ParseType(FILE_STRUCTURE[str(i)])
    if parse_type == ParseType.MIN_TWO_GLAS:
        data_phrases = list(
            filter_phrases_with_min_two_different_glas(data_phrases))
    elif parse_type == ParseType.CHET_LEN:
        data_phrases = list(
            filter(lambda phrase: len(phrase) % 2 == 0, data_phrases))
    elif parse_type == ParseType.USE_ALL:
        pass
    data_phrases.sort(reverse=True)
    output_json[str(i)] = data_phrases
    i += 1

with open("output.json", "w", encoding='utf-8') as f:
    json.dump(output_json, f)
```
